```python
# 라이브러리
import pymysql
import csv

import pandas as pd
import numpy as np
import re
```

```python
# 해당 엑셀 파일 Sheet2를 데이터프레임으로 불러오는 코드
df_Army_s2 = pd.read_excel('단어_의병_20211103_JH.xlsx', sheet_name='Sheet2')
df_Army_s2
```

```python
# 열값 'DEL'의 값이 D 인 경우만 출력 
is_D = df_Army_s2[df_Army_s2['DEL'] == 'D']
is_D
```

```python
is_place = df_Army_s2[df_Army_s2['SORT'] == '지명']
is_space = df_Army_s2[df_Army_s2['SORT'] == '장소']
is_name = df_Army_s2[df_Army_s2['SORT'] == '인명']
is_period = df_Army_s2[df_Army_s2['SORT'] == '시기']
is_job = df_Army_s2[df_Army_s2['SORT'] == '직위']
is_law = df_Army_s2[df_Army_s2['SORT'] == '법률']
is_penal = df_Army_s2[df_Army_s2['SORT'] == '수형']
is_group = df_Army_s2[df_Army_s2['SORT'] == '단체명']
```

```python
# 해당 엑셀 파일 Sheet1을 데이터 프레임으로 불러오는 것
# 주요 핵심 Sheet
df_Army_s1 = pd.read_excel('단어_의병_20211103_JH.xlsx', sheet_name='Sheet1')
df_Army_s1
```

```python
# 비어있는 리스트
df_del = []
# for문을 통해 중복되는 단어들은 제거하면서, 제거할 단어들의 리스트를 만든다. 
for value in is_D['DATA_CH'].tolist():
    if value not in df_del:
        df_del.append(value)
```

```python
df_sort = []
for value_1 in is_place['DATA_CH'].tolist():
    if value_1 not in df_sort:
        df_sort.append(value_1)

df_sort2 = []
for value_2 in is_space['DATA_CH'].tolist():
    if value_2 not in df_sort2:
        df_sort2.append(value_2)
        
df_sort3 = []
for value_3 in is_name['DATA_CH'].tolist():
    if value_3 not in df_sort3:
        df_sort3.append(value_3)
        
df_sort4 = []
for value_4 in is_period['DATA_CH'].tolist():
    if value_4 not in df_sort4:
        df_sort4.append(value_4)

df_sort5 = []
for value_5 in is_job['DATA_CH'].tolist():
    if value_5 not in df_sort5:
        df_sort5.append(value_5)
        
df_sort6 = []
for value_6 in is_law['DATA_CH'].tolist():
    if value_6 not in df_sort6:
        df_sort6.append(value_6)
        
df_sort7 = []
for value_7 in is_penal['DATA_CH'].tolist():
    if value_7 not in df_sort7:
        df_sort7.append(value_7)
        
df_sort8 = []
for value_8 in is_group['DATA_CH'].tolist():
    if value_8 not in df_sort8:
        df_sort8.append(value_8)
```

```python
# 단어 제거할 리스트의 갯수만큼의 횟수를 반복한다.
for w in range(0,len(df_del)):
    # DATA_CH 의 값이 df_sort 리스트의 값과 같을 경우 D를 삽입한다.
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_del[w],'DEL'] = 'D'
# df_Army_s1
```

```python
# place
for a in range(0,len(df_sort)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort[a],'SORT'] = '지명'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort[a],'DEL'] = 'S'
    
# space
for b in range(0,len(df_sort2)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort2[b],'SORT'] = '장소'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort2[b],'DEL'] = 'S'

# name
for c in range(0,len(df_sort3)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort3[c],'SORT'] = '인명'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort3[c],'DEL'] = 'S'
    
# period    
for d in range(0,len(df_sort4)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort4[d],'SORT'] = '시기'   
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort4[d],'DEL'] = 'S'
    
# job
for e in range(0,len(df_sort5)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort5[e],'SORT'] = '직위'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort5[e],'DEL'] = 'S'
    
# law    
for f in range(0,len(df_sort6)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort6[f],'SORT'] = '법률'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort6[f],'DEL'] = 'S'
    
# penal    
for g in range(0,len(df_sort7)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort7[g],'SORT'] = '수형'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort7[g],'DEL'] = 'S'
    
# group
for h in range(0,len(df_sort8)):
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort8[h],'SORT'] = '단체명'
    df_Army_s1.loc[df_Army_s1['DATA_CH'] == df_sort8[h],'DEL'] = 'S'
```
