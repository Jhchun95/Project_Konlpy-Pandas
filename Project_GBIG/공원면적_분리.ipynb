```python
# pandas : 데이타 분석(Data Analysis)을 위해 널리 사용되는 파이썬 라이브러리 패키지
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 프로젝트때 사용할 엑셀 데이터를 읽어온다.
대전_공원 = pd.read_csv('C:/Users/user/Desktop/프로젝트/대전시_공원.csv',encoding='cp949')
```

```python
# 변수명만 가져옴.
대전_공원.columns 
```

```python
# 원하는 변수만 가져와서 다시 선언하여 저장함
대전_공원 = 대전_공원[['관리번호','공원명','공원구분','소재지도로명주소','위도','경도','공원면적','구','EMD_CD','EMD_KOR_NM']] 
```

### 면적이 1500이하는 제외 

```python
# 면적이 1500~10000인 공원만 가져옴
대전공원_1500 = 대전_공원[(대전_공원['공원면적'] < 10000) & (대전_공원['공원면적'] >= 1500) ]
대전공원_1500['공원구분'].value_counts()
```

```python
# 면적이 10000~30000인 공원만 가져옴
대전공원_10000 = 대전_공원[(대전_공원['공원면적'] < 30000) &(대전_공원['공원면적'] >= 10000)]
대전공원_10000['공원구분'].value_counts()
```

```python
# 면적이 30000이상인 공원만 가져옴
대전공원_30000 = 대전_공원[(대전_공원['공원면적'] >= 30000)]
대전공원_30000['공원구분'].value_counts()
```

```python
#각각 csv로 저장
대전공원_1500.to_csv('C:/Users/user/Desktop/프로젝트/대전공원_1500.csv',encoding='cp949')
대전공원_10000.to_csv('C:/Users/user/Desktop/프로젝트/대전공원_10000.csv',encoding='cp949')
대전공원_30000.to_csv('C:/Users/user/Desktop/프로젝트/대전공원_30000.csv',encoding='cp949')
```
