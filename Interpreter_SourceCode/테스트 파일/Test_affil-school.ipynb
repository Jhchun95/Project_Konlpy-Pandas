```python
import pandas as pd
import re
```

```python
sm_area = pd.read_excel('공훈록 전체_JH.xlsx', sheet_name= "Sheet3")
sm_area
```

```python
sm_alist = sm_area['CONTENT'][0].split('.')
sm_list = []

# 문장에 '광주'라는 단어가 들어가있으면, 리스트에 하나씩 추가한다.
for x in sm_alist:
    if "광주" in x:
        sm_list.append(x)
sm_st = ','.join(sm_list)
sm_st = re.sub(r'[(\u4e00-\u9fff)]+', "" , sm_st)
m = re.findall('광주', sm_st)
sm_rarea = pd.DataFrame({     'MNG_NO' : sm_area['MNG_NO'][0],
                            'NAME_KO' : sm_area['NAME_KO'][0],
                            'WORKOUT_AFFIL' : sm_area['WORKOUT_AFFIL'][0],
                            'SCHOOL' : m })

for zz in range(1, sm_area['MNG_NO'].count()):
    
    # NAME_KO 두 번째 인물 출력
    # 위의 코드와 동일하고, 중복 제거하는 코드만 추가되어있다.
    sm_t_list2 = sm_area['CONTENT'][zz].split('.')
    sm_list2 = []

    for y in sm_t_list2:
        if "광주" in y:
            sm_list2.append(y)
    # print(sm_list2)
    # len(sm_list)

    sm_st2 = ','.join(sm_list2)
    sm_st2 = re.sub(r'[(\u4e00-\u9fff)]+', "" , sm_st2)
    m2 = re.findall('광주', sm_st2)

    # 리스트 내 중복된 단어 제거
    arr_m = []
    for v in m2:
        if v not in arr_m:
            arr_m.append(v)
    arr_m

    sm_rarea2 = pd.DataFrame({     'MNG_NO' : sm_area['MNG_NO'][zz],
                                'NAME_KO' : sm_area['NAME_KO'][zz],
                                'WORKOUT_AFFIL' : sm_area['WORKOUT_AFFIL'][zz],
                                'SCHOOL' : arr_m })
    sm_rarea = sm_rarea.append(sm_rarea2, ignore_index= True)
    sm_rarea
print(sm_rarea)  
```
