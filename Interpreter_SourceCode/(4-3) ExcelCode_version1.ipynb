```python
df_evelist = df_frame[['word']].values.tolist()
# df_evelist
```

```python
ListoStr = ''.join([str(element) for element in df_evelist[0] ])
mpva_evenouns = hannanum.nouns(ListoStr)
mpva_evenouns
```

```python
df_1row = df_frame[0:1]
df_1row
```

```python
# 형태소 분석기를 통해 나온 단어들로새로운 dataframe형을 만듬
df_evecols = pd.DataFrame(mpva_evenouns)
# column명 'Word'로 선언
df_evecols.columns = ['data_ch']
print(df_evecols)
```

```python
syn_df = pd.concat([df_1row, df_evecols], axis= 1).fillna(method='pad').drop('word', axis =1)
syn_df = syn_df.astype({'seq': 'int'}).astype({'asis_seq': 'int'})
syn_df
```

```python
syn_df["serial_seq"] = syn_df.index + 1
syn_df = syn_df[['seq', 'asis_seq', 'serial_seq', 'data_ch']]
syn_df
```
